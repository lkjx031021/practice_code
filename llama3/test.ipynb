```python
import torch

theta = 10000.0
dim = 512
max_position = 5000
batch_size = 2
seq_len = 128

x = torch.rand(batch_size, seq_len, dim)
```

```python
x.shape
```

```python
freqs = 1.0 / (theta ** (torch.arange(0, dim, 2)[: (dim // 2)].float() / dim))
m = torch.arange(max_position, device=freqs.device)
freqs = torch.outer(m, freqs).float()
pos_cis = torch.polar(torch.ones_like(freqs), freqs)
```

```python
print(freqs.shape)
print(freqs)
print(pos_cis.shape)
```
